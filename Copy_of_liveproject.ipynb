task 2- data cleaning


```python
# we create copy of the dataset

clean_df = df.copy()
```

```python
# the Convert column names to lowercase

clean_df.columns = clean_df.columns.str.lower()

#after that we Replace spaces with underscore

clean_df.columns = clean_df.columns.str.replace(" ", "_")

print(clean_df.columns)
```

```python
# Convert to proper case

clean_df["company_type"] = clean_df["company_type"].str.title()

print(clean_df["company_type"].value_counts())
```

```python
clean_df["current_ctc"] = clean_df["current_ctc"].str.replace("₹","")
clean_df["current_ctc"] = clean_df["current_ctc"].str.replace("LPA","")
clean_df["current_ctc"] = clean_df["current_ctc"].str.replace(",","")
clean_df["current_ctc"] = clean_df["current_ctc"].astype(float)
```

```python
clean_df["previous_ctc"] = clean_df["previous_ctc"].str.replace("₹","")
clean_df["previous_ctc"] = clean_df["previous_ctc"].str.replace("LPA","")
clean_df["previous_ctc"] = clean_df["previous_ctc"].str.replace(",","")

clean_df["previous_ctc"] = pd.to_numeric(
    clean_df["previous_ctc"],
    errors="coerce"
)
```

```python
clean_df = clean_df.drop_duplicates()

print(clean_df.shape)
```

## Task 3.1 – Salary Distribution by Role


```python
#for the list of column present
print(clean_df.columns)
```

```python
#This will show the actual column names
clean_df.rename(columns={"role_": "role"}, inplace=True)

print(clean_df.columns)
```

```python
# Median, Mean, Min and Max salary for each role
role_salary = clean_df.groupby("role")["current_ctc"].agg(
    ["median", "mean", "min", "max"]
)

role_salary = role_salary.sort_values(by="median", ascending=False)

print(role_salary)
```

```python
#to find highest paying tole and lowest paying role
print("Highest Paying Role")
print(role_salary.head(1))
```

```python
print("Lowest Paying Role")
print(role_salary.tail(1))
```

task 3.2-Salary Based on Experience

```python
# Select only Backend employees

backend = clean_df[clean_df["role"]=="SDE Backend"]

# Create experience groups

backend["experience"] = pd.cut(
    backend["years_exp"],
    bins=[0,1,3,5,100],
    labels=["0-1","2-3","4-5","6+"]
)

# Find median salary

print(backend.groupby("experience")["current_ctc"].median())
```

task 3.3-Skill Premium

```python
skills = ["AWS","ML","System Design","Kubernetes"]
# List of skills to analyze
# for Loop for each skill one by one

for skill in skills:

    yes = clean_df[
        clean_df["skills"].fillna("").str.contains(skill)
    ]["current_ctc"].median()

    no = clean_df[
        ~clean_df["skills"].fillna("").str.contains(skill)
    ]["current_ctc"].median()

    print(skill)
    print("With Skill :", yes)
    print("Without Skill :", no)
    print()
```

task 3.4-compare salary

```python
backend = clean_df[clean_df["role"]=="SDE Backend"]

print(
    backend.groupby("company_type")["current_ctc"].median()
)
```

```python
# Create experience groups

clean_df["experience"] = pd.cut(
    clean_df["years_exp"],
    bins=[0,1,3,5,100],
    labels=["0-1","2-3","4-5","6+"]
)

# Median salary of each group

clean_df["group_salary"] = clean_df.groupby(
    ["role","company_type","experience"]
)["current_ctc"].transform("median")

# Salary difference

clean_df["gap"] = clean_df["current_ctc"] - clean_df["group_salary"]

# Show lowest paid employees

underpaid = clean_df.sort_values("gap")

print(
    underpaid[
        ["employee_id","role","current_ctc","gap"]
    ].head(10)
)
```
